```python
import nltk

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')

from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer


text = """
Bluetooth is a short-range wireless technology standard that is used for exchanging data
between fixed and mobile devices. Google and Microsoft use wireless technology.
India is developing AI rapidly.
"""

print("Original Text:")
print(text)
print("-" * 50)

sentences = nltk.sent_tokenize(text)

words = nltk.word_tokenize(text)

print("Word Tokens:")
print(words)
print("-" * 50)

stop_words = set(stopwords.words("english"))

filtered_words = [
    word for word in words
    if word.lower() not in stop_words and word.isalpha()
]

print("After Stop Word Removal:")
print(filtered_words)
print("-" * 50)


stemmer = PorterStemmer()

stemmed_words = [
    stemmer.stem(word) for word in filtered_words
]

print("After Stemming:")
print(stemmed_words)
print("-" * 50)



lemmatizer = WordNetLemmatizer()

lemmatized_words = [
    lemmatizer.lemmatize(word) for word in filtered_words
]

print("After Lemmatization:")
print(lemmatized_words)
print("-" * 50)



pos_tags = nltk.pos_tag(filtered_words)

print("POS Tagging:")
print(pos_tags)
print("-" * 50)



named_entities = nltk.ne_chunk(pos_tags)

print("Named Entity Recognition:")
print(named_entities)

```
