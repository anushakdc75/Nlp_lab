```python
sentences = [s.strip() for s in sentences if s.strip()]
```

```python
tokenized_sentences = [sentence.split() for sentence in sentences]
```

```python
N = len(tokenized_sentences)
print("Total Sentences:", N)
```

```python
for i, sentence in enumerate(tokenized_sentences):
    print("\nSentence", i+1)

    total_words = len(sentence)
    word_count = Counter(sentence)

    for word in word_count:

        # TF
        tf = word_count[word] / total_words
        print (tf)
```

```python
n = sum(1 for s in tokenized_sentences if word in s)
```

```python
idf = math.log(N / n)
tf_idf = tf * idf
```

```python
print(f"{word} → TF:{tf:.3f}, IDF:{idf:.3f}, TF-IDF:{tf_idf:.3f}")
```
